```python
df = pd.read_csv('delivery_orders_march.csv')
```

```python
final_output ={}

df = df.applymap(lambda s:s.lower() if type(s) == str else s)
```

```python
temp_dict = {}

for index,row in df.iterrows():
    temp_dict[df["orderid"][index]] = []
    if "manila" in df["buyeraddress"][index]:
        temp_dict[df["orderid"][index]].append("manila")
    elif "luzon" in df["buyeraddress"][index]:
         temp_dict[df["orderid"][index]].append("luzon")
    elif "visayas" in df["buyeraddress"][index]:
         temp_dict[df["orderid"][index]].append("visayas")
    elif "mindanao" in df["buyeraddress"][index]:
         temp_dict[df["orderid"][index]].append("mindanao")
            
    if "manila" in df["selleraddress"][index]:
        temp_dict[df["orderid"][index]].append("manila")
    elif "luzon" in df["selleraddress"][index]:
         temp_dict[df["orderid"][index]].append("luzon")
    elif "visayas" in df["selleraddress"][index]:
         temp_dict[df["orderid"][index]].append("visayas")
    elif "mindanao" in df["selleraddress"][index]:
         temp_dict[df["orderid"][index]].append("mindanao")
            
    if pd.isnull(df["2nd_deliver_attempt"][index]):
        temp_dict[df["orderid"][index]].append(0)
    else:
        temp_dict[df["orderid"][index]].append(1)
        
        
    if index % 100000 == 0:
        print(index)
        
    
```

```python
import datetime as dt
import math
def checksunandph(date1,date2):
    checkingno = 0 
    
    #check for sunday
    d1 = date1.day
    d2 = date2.day
    
    m1 = date1.month
    m2 = date2.month
    
    diff = 0
    
    check = 0 

    
    while check == 0 :
        if m1 == m2:
            if d1 == d2:
                check = 1
            else:
                 diff += 1
                 d1 += 1
                
        else:
            if d1 == d2 :
                 check = 1
            else:
                diff +=1
                d1 += 1
        
                try :
                    (dt.datetime(2020,m1,d1))
                except ValueError :
                    d1 = 1 
                 
            
 
    start= date1.weekday()
    ending = start + diff 
    if ending > 6:
        sundays = math.floor(ending/6)
        ##print(sundays)
        checkingno -= sundays 
        
    #check for PH
    
    first = dt.datetime(2020,3,25)
    second =dt.datetime(2020,3,30)
    third = dt.datetime(2020,3,31)
    
    if date1< first < date2:
        checkingno -= 1
    if  date1< second < date2:
        checkingno -= 1
    if  date1< third < date2:
        checkingno -= 1 
    
   
        
    return (diff + checkingno)
```

```python
for index,row in df.iterrows():
    start = dt.datetime.fromtimestamp(df["pick"][index])
    try1 = dt.datetime.fromtimestamp(df["1st_deliver_attempt"][index])
    days = checksunandph(start,try1)
    temp_dict[df["orderid"][index]].append(days)
    
    if temp_dict[df["orderid"][index]][2] == 1:
        try2 = dt.datetime.fromtimestamp(df["2nd_deliver_attempt"][index])
        days1 = checksunandph(try1,try2)
        temp_dict[df["orderid"][index]].append(days1)
    else:
        temp_dict[df["orderid"][index]].append(0)

    if index %100000== 0:
        print(index)
```
